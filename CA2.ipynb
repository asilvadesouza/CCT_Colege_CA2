```python
# ============================================================
# 1. IMPORTAÇÃO DAS BIBLIOTECAS
# ============================================================

# Objetivo:
# Importar as bibliotecas necessárias para leitura, limpeza,
# análise exploratória e visualização dos dados.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração visual básica dos gráficos
plt.rcParams["figure.figsize"] = (10, 5)
sns.set_theme(style="whitegrid")
```

```python
# ============================================================
# 2. LEITURA DO DATASET
# ============================================================

# Objetivo:
# Carregar o arquivo CSV da FAOSTAT para dentro de um DataFrame.

df = pd.read_csv("FAOSTAT_data_en_5-24-2026.csv")

# Visualizar as primeiras linhas
df.head()
```

```python
# ============================================================
# 3. INSPEÇÃO INICIAL DOS DADOS
# ============================================================

# Objetivo:
# Verificar o tamanho do dataset, tipos de dados,
# valores ausentes e principais categorias existentes.

print("Shape do dataset:", df.shape)

print("\nInformações gerais:")
print(df.info())

print("\nValores ausentes por coluna:")
print(df.isnull().sum())

print("\nPaíses disponíveis:")
print(df["Area"].unique())

print("\nCategorias do sistema alimentar:")
print(df["Item"].unique())

print("\nTipos de energia:")
print(df["Element"].unique())

print("\nDescrição das flags:")
print(df["Flag Description"].value_counts())
```

```python
# ============================================================
# 4. PADRONIZAÇÃO DOS NOMES DAS COLUNAS
# ============================================================

# Objetivo:
# Padronizar os nomes das colunas para facilitar o uso no Python.
# Exemplo: "Flag Description" vira "flag_description".

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)

df.head()
```

```python
# ============================================================
# 5. VERIFICAÇÃO DE DUPLICADOS
# ============================================================

# Objetivo:
# Identificar se existem linhas duplicadas no dataset.

duplicates = df.duplicated().sum()

print("Número de linhas duplicadas:", duplicates)
```

```python
# Caso existam duplicados, eles podem ser removidos.
# Se o resultado anterior for 0, este comando não altera o dataset.

df = df.drop_duplicates()

print("Shape após remoção de duplicados:", df.shape)
```

```python
# ============================================================
# 6. ESTATÍSTICAS DESCRITIVAS
# ============================================================

# Objetivo:
# Avaliar média, desvio padrão, mínimo, máximo e quartis
# da variável numérica principal: value.

df["value"].describe()
```

```python
# ============================================================
# 7. FREQUÊNCIA DAS CATEGORIAS
# ============================================================

# Objetivo:
# Verificar a distribuição dos registros por país,
# categoria alimentar e tipo de energia.

print("Registros por país:")
print(df["area"].value_counts())

print("\nRegistros por categoria alimentar:")
print(df["item"].value_counts())

print("\nRegistros por tipo de energia:")
print(df["element"].value_counts())
```

```python
# ============================================================
# 7. FREQUÊNCIA DAS CATEGORIAS
# ============================================================

# Objetivo:
# Verificar a distribuição dos registros por país,
# categoria alimentar e tipo de energia.

print("Registros por país:")
print(df["area"].value_counts())

print("\nRegistros por categoria alimentar:")
print(df["item"].value_counts())

print("\nRegistros por tipo de energia:")
print(df["element"].value_counts())
```

```python
# ============================================================
# 8. HISTOGRAMA DA VARIÁVEL VALUE
# ============================================================

# Objetivo:
# Avaliar a distribuição dos valores de consumo energético.
# Isso ajuda a identificar assimetria, concentração de valores
# pequenos e possíveis valores extremos.

sns.histplot(df["value"], bins=30, kde=True)

plt.title("Distribution of Energy Use Values")
plt.xlabel("Energy Use (TJ)")
plt.ylabel("Frequency")
plt.show()
```

```python
# ============================================================
# 9. BOXPLOT PARA IDENTIFICAÇÃO DE OUTLIERS
# ============================================================

# Objetivo:
# Verificar se existem valores extremos por categoria do sistema alimentar.

sns.boxplot(data=df, x="item", y="value")

plt.title("Energy Use Distribution by Food System Activity")
plt.xlabel("Food System Activity")
plt.ylabel("Energy Use (TJ)")
plt.xticks(rotation=45)
plt.show()
```

```python
# ============================================================
# 10. CONSUMO MÉDIO POR PAÍS AO LONGO DO TEMPO
# ============================================================

# Objetivo:
# Comparar a evolução do consumo energético médio entre os países
# de 2013 a 2023.

sns.lineplot(
    data=df,
    x="year",
    y="value",
    hue="area",
    estimator="mean",
    errorbar=None
)

plt.title("Average Energy Use by Country, 2013–2023")
plt.xlabel("Year")
plt.ylabel("Average Energy Use (TJ)")
plt.legend(title="Country")
plt.show()
```

```python
# ============================================================
# 11. ANÁLISE ESPECÍFICA DA IRLANDA
# ============================================================

# Objetivo:
# Filtrar apenas os dados da Irlanda para analisar seu comportamento
# em cada atividade do sistema alimentar.

ireland = df[df["area"] == "Ireland"]

sns.lineplot(
    data=ireland,
    x="year",
    y="value",
    hue="item",
    estimator="sum",
    errorbar=None
)

plt.title("Ireland: Energy Use by Food System Activity, 2013–2023")
plt.xlabel("Year")
plt.ylabel("Energy Use (TJ)")
plt.legend(title="Food System Activity")
plt.show()
```

```python
# ============================================================
# 12. IRLANDA: CONSUMO POR TIPO DE ENERGIA
# ============================================================

# Objetivo:
# Identificar quais fontes de energia predominam no sistema alimentar
# da Irlanda.

sns.barplot(
    data=ireland,
    x="item",
    y="value",
    hue="element",
    estimator="sum",
    errorbar=None
)

plt.title("Ireland: Energy Use by Activity and Energy Source")
plt.xlabel("Food System Activity")
plt.ylabel("Total Energy Use (TJ)")
plt.xticks(rotation=45)
plt.legend(title="Energy Source")
plt.show()
```

```python
# ============================================================
# 13. VERIFICAÇÃO DE COMBINAÇÕES ITEM × ELEMENT
# ============================================================

# Objetivo:
# Mesmo sem valores nulos explícitos, algumas combinações entre
# atividade alimentar e tipo de energia podem não existir.
# Esta análise identifica a disponibilidade de registros.

pivot_count = pd.pivot_table(
    df,
    index="item",
    columns="element",
    values="value",
    aggfunc="count"
)

pivot_count
```

```python
# ============================================================
# 14. HEATMAP DE DISPONIBILIDADE DOS DADOS
# ============================================================

# Objetivo:
# Visualizar quais combinações de item e energia têm mais ou menos registros.
# Isso ajuda a identificar ausência estrutural de dados.

sns.heatmap(
    pivot_count,
    annot=True,
    fmt=".0f",
    cmap="Blues"
)

plt.title("Availability of Observations by Activity and Energy Source")
plt.xlabel("Energy Source")
plt.ylabel("Food System Activity")
plt.show()
```

```python
# ============================================================
# 15. COMPARAÇÃO IRLANDA × MÉDIA DOS OUTROS PAÍSES
# ============================================================

# Objetivo:
# Comparar o consumo energético total da Irlanda com a média
# dos demais países selecionados.

total_by_country_year = (
    df.groupby(["area", "year"])["value"]
    .sum()
    .reset_index()
)

ireland_total = total_by_country_year[
    total_by_country_year["area"] == "Ireland"
].copy()

europe_avg = (
    total_by_country_year[
        total_by_country_year["area"] != "Ireland"
    ]
    .groupby("year")["value"]
    .mean()
    .reset_index()
)

europe_avg["area"] = "Average of Selected Countries"

comparison = pd.concat([ireland_total, europe_avg])

sns.lineplot(
    data=comparison,
    x="year",
    y="value",
    hue="area",
    marker="o"
)

plt.title("Ireland vs Average of Selected Countries")
plt.xlabel("Year")
plt.ylabel("Total Energy Use (TJ)")
plt.legend(title="Comparison")
plt.show()
```
