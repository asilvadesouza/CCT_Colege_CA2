```python
# ============================================================
# AGGREGATE ENERGY CONSUMPTION BY COUNTRY AND YEAR USING SQL
# ============================================================
#
# Purpose:
# Calculate the total energy consumption for each country
# and year directly from the SQLite database.
#
# This aggregation demonstrates the use of SQL functions
# such as SUM(), GROUP BY, and ORDER BY to summarize
# data efficiently before analysis and visualization.
#
# Database:
# - faostat_food_energy.db
#
# Table:
# - food_energy
#
# Output:
# - Total energy consumption by country and year
#
# ============================================================

# SQL query to aggregate energy consumption by country and year
query = """
SELECT
    area,
    year,
    SUM(value) AS total_energy
FROM food_energy
GROUP BY area, year
ORDER BY area, year;
"""

# Execute the SQL query and load the aggregated results
# into a Pandas DataFrame
country_year_energy_sql = pd.read_sql_query(query, conn)

# Display the first records of the aggregated dataset
country_year_energy_sql.head()
```

```python
# ============================================================
# SQL QUERY PERFORMANCE TEST
# ============================================================
#
# Purpose:
# Measure the execution time of a SQL query performed
# on the SQLite database.
#
# This test evaluates the efficiency of data retrieval
# and aggregation operations executed directly within
# the database engine.
#
# Query Operation:
# - Aggregate total energy consumption by country and year
# - Group results using SQL GROUP BY
#
# Performance Metric:
# - Execution time in seconds
#
# ============================================================

# Record the start time
start = time.time()

# SQL query to aggregate energy consumption
query = """
SELECT
    area,
    year,
    SUM(value) AS total_energy
FROM food_energy
GROUP BY area, year;
"""

# Execute the query and load the results into a DataFrame
test_query = pd.read_sql_query(query, conn)

# Record the end time
end = time.time()

# Calculate and display execution time
print("SQL query execution time:", end - start, "seconds")
```

```python
# ============================================================
# RECORD COUNT BY YEAR USING SQL
# ============================================================
#
# Purpose:
# Retrieve the number of observations available for each
# year directly from the SQLite database.
#
# This query is used to evaluate the temporal distribution
# of the dataset and verify whether records are evenly
# distributed across the study period (2013–2023).
#
# SQL Operations:
# - COUNT(): count observations
# - GROUP BY: aggregate records by year
# - ORDER BY: sort results chronologically
#
# ============================================================

# SQL query to count records by year
query = """
SELECT
    year,
    COUNT(*) AS count
FROM food_energy
GROUP BY year
ORDER BY year;
"""

# Execute the query and load results into a DataFrame
records_year = pd.read_sql_query(query, conn)

# Display aggregated results
print(records_year)

# ============================================================
# VISUALIZATION
# ============================================================
#
# Purpose:
# Visualize the number of observations available for
# each year in the dataset.
#
# ============================================================

plt.figure(figsize=(8, 4))

sns.barplot(
    data=records_year,
    x="year",
    y="count"
)

plt.title("Records per Year")
plt.xlabel("Year")
plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# ============================================================
# RECORD COUNT BY COUNTRY USING SQL
# ============================================================
#
# Purpose:
# Retrieve the number of observations available for each
# country directly from the SQLite database.
#
# This query is used to evaluate the geographical
# distribution of the dataset and verify whether all
# countries are represented consistently throughout
# the analysis.
#
# SQL Operations:
# - COUNT(): count observations
# - GROUP BY: aggregate records by country
# - ORDER BY: sort countries by record count
#
# ============================================================

# SQL query to count records by country
query = """
SELECT
    area,
    COUNT(*) AS count
FROM food_energy
GROUP BY area
ORDER BY count DESC;
"""

# Execute the query and load results into a DataFrame
records_country = pd.read_sql_query(query, conn)

# Display aggregated results
print(records_country)

# ============================================================
# VISUALIZATION
# ============================================================
#
# Purpose:
# Visualize the number of observations available for
# each country in the dataset.
#
# ============================================================

plt.figure(figsize=(8, 4))

sns.barplot(
    data=records_country,
    x="area",
    y="count"
)

plt.title("Records per Country")
plt.xlabel("Country")
plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# ============================================================
# RECORD COUNT BY FOOD SYSTEM ACTIVITY USING SQL
# ============================================================
#
# Purpose:
# Retrieve the number of observations available for each
# food system activity directly from the SQLite database.
#
# This query is used to evaluate the distribution of
# records across the different food system activities
# included in the study.
#
# SQL Operations:
# - COUNT(): count observations
# - GROUP BY: aggregate records by activity
# - ORDER BY: sort activities by record count
#
# ============================================================

# SQL query to count records by food system activity
query = """
SELECT
    item,
    COUNT(*) AS count
FROM food_energy
GROUP BY item
ORDER BY count DESC;
"""

# Execute the query and load results into a DataFrame
records_item = pd.read_sql_query(query, conn)

# Display aggregated results
print(records_item)

# ============================================================
# VISUALIZATION
# ============================================================
#
# Purpose:
# Visualize the number of observations available for
# each food system activity.
#
# ============================================================

plt.figure(figsize=(8, 4))

sns.barplot(
    data=records_item,
    x="item",
    y="count"
)

plt.title("Records per Food System Activity")
plt.xlabel("Food System Activity")
plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# ============================================================
# RECORD COUNT BY ENERGY SOURCE USING SQL
# ============================================================
#
# Purpose:
# Retrieve the number of observations available for each
# energy source directly from the SQLite database.
#
# This query is used to evaluate the distribution of
# records across the different energy sources included
# in the dataset.
#
# SQL Operations:
# - COUNT(): count observations
# - GROUP BY: aggregate records by energy source
# - ORDER BY: sort energy sources by record count
#
# ============================================================

# SQL query to count records by energy source
query = """
SELECT
    element,
    COUNT(*) AS count
FROM food_energy
GROUP BY element
ORDER BY count DESC;
"""

# Execute the query and load results into a DataFrame
records_element = pd.read_sql_query(query, conn)

# Display aggregated results
print(records_element)

# ============================================================
# VISUALIZATION
# ============================================================
#
# Purpose:
# Visualize the number of observations available for
# each energy source in the dataset.
#
# ============================================================

plt.figure(figsize=(8, 4))

sns.barplot(
    data=records_element,
    x="element",
    y="count"
)

plt.title("Records per Energy Source")
plt.xlabel("Energy Source")
plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# ============================================================
# AVERAGE ENERGY CONSUMPTION BY COUNTRY USING SQL
# ============================================================
#
# Purpose:
# Calculate the average energy consumption for each
# country directly from the SQLite database.
#
# This query is used to compare average energy
# consumption levels across countries and identify
# geographical differences within the dataset.
#
# SQL Operations:
# - AVG(): calculate average energy consumption
# - GROUP BY: aggregate records by country
# - ORDER BY: sort countries by average consumption
#
# Output:
# - Mean energy consumption (TJ) for each country
#
# ============================================================

# SQL query to calculate average energy consumption by country
query = """
SELECT
    area,
    AVG(value) AS mean_energy
FROM food_energy
GROUP BY area
ORDER BY mean_energy;
"""

# Execute the query and load results into a DataFrame
country_mean = pd.read_sql_query(query, conn)

# Display aggregated results
print(country_mean)

# ============================================================
# VISUALIZATION
# ============================================================
#
# Purpose:
# Visualize the average energy consumption for each
# country included in the study.
#
# ============================================================

plt.figure(figsize=(8, 5))

plt.scatter(
    country_mean["area"],
    country_mean["mean_energy"]
)

plt.title("Average Energy Consumption by Country")
plt.xlabel("Country")
plt.ylabel("Average Energy Consumption (TJ)")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# ============================================================
# TOTAL ENERGY CONSUMPTION BY COUNTRY AND YEAR USING SQL
# ============================================================
#
# Purpose:
# Calculate the total energy consumption for each
# country and year directly from the SQLite database.
#
# This query is used to analyse temporal energy
# consumption trends and compare how energy use
# evolved across countries during the study period.
#
# SQL Operations:
# - SUM(): calculate total energy consumption
# - GROUP BY: aggregate records by country and year
# - ORDER BY: sort results chronologically within
#   each country
#
# Output:
# - Total energy consumption (TJ) by country and year
#
# ============================================================

# SQL query to aggregate energy consumption by country and year
query = """
SELECT
    area,
    year,
    SUM(value) AS total_energy
FROM food_energy
GROUP BY area, year
ORDER BY area, year;
"""

# Execute the query and load results into a DataFrame
country_year_energy = pd.read_sql_query(query, conn)

# Display the first records of the aggregated dataset
print(country_year_energy.head())
```

The confidence interval was used to estimate the population mean of energy consumption associated with the Irish food system. While the sample mean provides only a point value, the confidence interval allows incorporating the statistical uncertainty associated with the observed data.

Theoretical Basis

In statistical studies, it is rarely possible to observe the entire population. Thus, a sample is used to estimate population parameters. The confidence interval represents a range of plausible values ​​for the parameter of interest (Montgomery and Runger, 2018).

Formula:

CI = x̄ ± tα/2 × (s/√n)

Where:

x̄ = sample mean;

s = standard deviation;

n = number of observations;

t = critical value of the Student's t-distribution.

```python
# ============================================================
# CONFIDENCE INTERVAL FOR IRELAND
# ============================================================
#
# Purpose:
# Calculate the 95% confidence interval for Ireland's
# energy consumption values.
#
# The confidence interval provides a range of values
# within which the true population mean is likely
# to be found.
#
# ============================================================

import scipy.stats as stats

# Select energy consumption values for Ireland
ireland_values = df[df["area"] == "Ireland"]["value"]

# Calculate the sample mean
mean_ireland = ireland_values.mean()

# Calculate the standard error of the mean
sem = stats.sem(ireland_values)

# Calculate the 95% confidence interval
ci_ireland = stats.t.interval(
    0.95,
    df=len(ireland_values) - 1,
    loc=mean_ireland,
    scale=sem
)

print(ci_ireland)
```

The 95% confidence interval estimated that the average energy consumption of the Irish food system is between 3405.56 TJ and 5117.45 TJ. This result provides a measure of precision for the observed average and reduces the uncertainty associated with international comparisons made in this study. . However, the interval does not explain the causes of the observed differences nor does it identify relationships between variables. Related studies using FAO agricultural data and agricultural decision support systems also employ confidence intervals to support comparisons between countries and validation of energy and production indicators (James et al., 2021; Osinga et al., 2022). For this work, the result establishes a solid statistical benchmark for comparing Ireland with other European Union countries and supporting future forecasting and machine learning steps.

Formula:

As a limitation, the test only compares global averages and does not consider possible specific differences between dietary activities or energy sources.

**5. One-Way ANOVA**

Objective

To simultaneously compare the seven countries analyzed.

Theoretical Framework

ANOVA allows us to verify if there are statistically significant differences between multiple groups.

Hypotheses:

H₀: All means are equal.

H₁: At least one mean is different.

Formula:

F = MSbetween / MSwithin

```python
groups = [
    group["value"].values
    for _, group in df.groupby("area")
]

anova_test = stats.f_oneway(*groups)

print(anova_test)
```

Analysis of variance (ANOVA) was used to simultaneously compare the average energy consumption levels of the seven countries analyzed: Ireland, Portugal, Spain, France, Germany, Italy, and Greece. While t-tests are suitable for comparisons between two groups, ANOVA allows for the simultaneous assessment of differences between multiple populations, reducing the risk of errors associated with performing multiple independent comparisons (Montgomery and Runger, 2018).

The result obtained showed F = 46.13 and p-value = 5.31 × 10⁻⁵⁰, a value significantly lower than the significance level of 0.05. Therefore, the null hypothesis was rejected, indicating that there are statistically significant differences between the average energy consumption levels observed in the analyzed countries.

This result confirms the evidence observed during the Exploratory Data Analysis. The time trend graphs demonstrated that Germany, France, and Italy have substantially higher average energy consumption levels than those observed in Ireland, Portugal, and Greece. ANOVA provides formal statistical validation of these differences.

The main advantage of this test is that it allows a global comparison between all the countries analyzed. However, ANOVA does not specifically identify which countries differ from each other; it only indicates that at least one group exhibits distinct behavior. For this reason, complementary tests or specific analyses between countries may be necessary to deepen the interpretation of the results.

Related studies in agricultural analysis, energy consumption, and food systems frequently use ANOVA to identify differences between regions or countries before implementing predictive models and decision support systems (James et al., 2021; Osinga et al., 2022).

In the context of this work, the result suggests that the energy consumption of the food system is not homogeneous among the European countries analyzed. This conclusion reinforces the importance of investigating specific factors associated with each country and provides support for the use of geographic and national variables in the Machine Learning models developed subsequently.

H₀: All countries have the same average energy consumption.

Statistical Decision

Since p-value = 5.31 × 10⁻⁵⁰ < 0.05, H₀ is rejected.

Conclusion

There are statistically significant differences between the average energy consumption levels of the countries analyzed. **6. Training and Testing Separation**

The data was split into training and testing sets to evaluate the models' ability to generalize to unseen data. An 80% training and 20% testing split was used. This step is essential to avoid evaluating the model only on the data used during training.

➡️ How training and testing separation works

Training/testing divides the dataset into two parts:

✅ 80% → training

✅ 20% → testing

The model learns patterns using only the training data. Then, it is evaluated on the test data, which it has not yet seen. This allows us to verify if the model has truly learned general patterns or has only "memorized" the data.

What each variable means:

* X = features: area, year, item, element
* y = target: value
* test_size=0.20 = 20% for testing
* random_state=42 = ensures that the result is reproducible

These features were used to predict the energy consumption of the food system; the model needs to be tested on data not used in the training. This helps to assess whether it can predict new cases, for example, new years, countries, or combinations of dietary activity and energy source.

```python
# ============================================================
# 3. DIVISÃO ENTRE TREINO E TESTE
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Treino:", X_train.shape)
print("Teste:", X_test.shape)
```

The separation between training and test data was applied to evaluate the generalization capacity of the predictive models. The training set was used to adjust the algorithm parameters, while the test set was reserved exclusively for evaluating the final performance. This approach reduces the risk of overfitting, a situation in which the model performs well on the training data but has a low capacity to predict unobserved data. In this study, an 80% split for training and 20% for testing was used, allowing the models to be evaluated more realistically in the problem of predicting the energy consumption of the food system.

**7. Preprocessing**

Since the dataset contains categorical variables, One-Hot Encoding was applied to the area, item, and element columns. The year variable was kept as numeric. For Linear Regression, standardization can improve model stability. For tree-based models, such as Random Forest and Gradient Boosting, the data scale is not as critical, but the same pipeline was maintained for methodological consistency.

```python
# ============================================================
# 4. PRÉ-PROCESSAMENTO
# ============================================================

# Separação entre Variáveis Categóricas e Numéricas
categorical_features = ["area", "item", "element"]
numeric_features = ["year"]



preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),#Codificação das Variáveis Categóricas
        ("num", StandardScaler(), numeric_features) # Normalização da Variável Temporal
    ]
)
```

**9 Modelo 2 — Random Forest com GridSearchCV**

Random Forest was chosen because it combines multiple decision trees, reducing variance and improving predictive capacity. This model is appropriate for the dataset because statistical tests indicated non-normality and outliers. Furthermore, Random Forest is able to capture non-linear relationships between country, year, dietary activity, and energy source.

```python
# ============================================================
# 6. MODELO 2: RANDOM FOREST COM GRIDSEARCHCV
# ============================================================

rf_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", RandomForestRegressor(random_state=42))
    ]
)

rf_param_grid = {
    "model__n_estimators": [100, 200],
    "model__max_depth": [None, 5, 10],
    "model__min_samples_split": [2, 5],
    "model__min_samples_leaf": [1, 2]
}

rf_grid = GridSearchCV(
    rf_pipeline,
    rf_param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1
)

rf_grid.fit(X_train, y_train)

print("Best parameters:", rf_grid.best_params_)
print("Best CV score:", rf_grid.best_score_)

y_pred_rf = rf_grid.predict(X_test)

rf_results = {
    "Model": "Random Forest",
    "MAE": mean_absolute_error(y_test, y_pred_rf),
    "RMSE": np.sqrt(mean_squared_error(y_test, y_pred_rf)),
    "R2": r2_score(y_test, y_pred_rf)
}

rf_results
```

**11 Comparison of the models**

The models were evaluated using three metrics: MAE, RMSE, and R². MAE measures the mean absolute error, RMSE penalizes larger errors, and R² indicates the proportion of variance explained by the model. Comparing these metrics allows us to identify which model presents the best balance between precision and explanatory power.
