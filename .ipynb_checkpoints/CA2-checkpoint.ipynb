Data Preparation & Visualization Tasks

1. Acquisition of Raw Data

 - The data used in this study were obtained from the FAOSTAT database, maintained by the Food and Agriculture Organization of the United Nations. FAOSTAT was chosen because it provides free access to agricultural and food data for more than 245 countries and territories, with historical series since 1961, allowing for consistent international comparisons.

 - The selected dataset belongs to the domain "Emissions from pre and post agricultural production," focusing on energy use in food system activities. Data from 2013 to 2023 were selected for the following countries:

 * Ireland
 * France
 * Germany
 * Greece
 * Italy
 * Portugal
 * Spain

The main variables of the dataset include:

 * Food Processing
 * Food Transport
 * Food Packaging
 * Food Retail
 * Food Household Consumption

These variables were selected because they represent different stages of the agri-food system, from industrial processing to household consumption.

 2. License and Permissions

 - According to the FAO's statistical database licensing policy, the standard license applied to corporate databases is Creative Commons Attribution-NonCommercial-ShareAlike 3.0 IGO — CC BY-NC-SA 3.0 IGO. This license allows copying, redistribution, and adaptation of the data for non-commercial purposes, provided the source is correctly attributed and derivative works are shared under the same terms.

* This has important implications for the project:

* The dataset can be used for academic purposes;

* FAO must be correctly cited;
* The use must be non-commercial;
* Any adaptation must respect the same license.

 3. Positive aspects of the acquisition

 - The main positive aspect is that FAOSTAT is an official, international, and recognized source. This increases the reliability of the data and allows comparisons between countries using standardized methodology. Furthermore, the data was already available in CSV format, facilitating import into Python/Jupyter Notebook.

 4. Negative Points of the Acquisition

 - The main negative point is that all records are marked as “Estimated value.” This means that the values ​​were estimated and not necessarily measured directly. Therefore, the results should be interpreted with caution. Another problem identified was the absence of some combinations between country, year, item, and energy type, which requires treatment during the data preparation stage.

 5. Evaluation of Raw Data and EDA

 - Exploratory Data Analysis, or EDA, was used to identify patterns, inconsistencies, missing values, possible outliers, and structural limitations of the dataset. This step is essential before modeling, as it allows understanding the data quality and justifying cleaning and transformation decisions.

The analyzed dataset contains:

 * Countries: 7

 * Unit: Terajoules (TJ)

The main columns are:

 * Area

 * Year

 * Unit

 * Flag

 * Flag Description

Initial Data Inspection

The first stage of the analysis consisted of a structural inspection of the dataset using the functions `shape()`, `info()`, `isnull().sum()`, `unique()`, and `value_counts()`. The objective was to understand the size of the dataset, the types of variables, the presence of missing values, and the composition of the main categories.

The dataset contains 1006 observations and 14 variables, covering energy consumption data in food system activities for seven European countries: France, Germany, Greece, Ireland, Italy, Portugal, and Spain. The variables include country, year, item analyzed, energy type, unit of measurement, numerical value, and data quality indicator.

The `info()` function showed that there are no explicit null values, as all 14 columns have 1006 non-null values. The `isnull().sum()` function confirmed this result, showing zero missing values ​​in all columns. However, this does not mean that all possible combinations of country, year, item, and energy type are available. Therefore, further analysis of expected combinations will be necessary to identify possible structural absences. The categories identified in the Item column were:

Food Processing

Food Transport

Food Packaging

Food Retail

Food Household Consumption

The Element column presents four types of energy consumption:

Energy Use (Natural Gas, including LNG)

Energy Use (Coal)

Energy Use (Electricity)

Energy Use (Heat)

The Value variable is in float64 format, indicating that it can be used directly in numerical analyses and machine learning models. Furthermore, the Year variable is in integer format, covering the period from 2013 to 2023.

An important point identified was that all records have the quality description "Estimated value". This means that the data were estimated by FAOSTAT and not necessarily measured directly in each country. This characteristic does not invalidate the analysis, but represents an important methodological limitation. For this reason, the results should be interpreted as official statistical estimates, and not as absolute measurements.

**Initial inspection indicates that the dataset is complete at the column level, with no explicit missing values. However, all records are classified as estimated values, which introduces uncertainty into the analysis. Therefore, further exploratory analysis is needed to assess data completeness in the country-year-item-energy type combinations and identify potential structural gaps, outliers, and inconsistencies before applying machine learning techniques.**

```python
# ============================================================
# COLUMN NAME STANDARDIZATION
# ============================================================
#
# Purpose:
# Standardize column names to improve readability and
# simplify data manipulation throughout the project.
#
# Transformations:
# - Remove leading and trailing spaces
# - Convert all characters to lowercase
# - Replace spaces with underscores
# - Remove parentheses
#
# Example:
# "Flag Description" -> "flag_description"
#
# ============================================================

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)

# Display the first records to verify that the
# column names were standardized successfully.
df.head()
```

```python
# ============================================================
# DUPLICATE RECORD DETECTION
# ============================================================
#
# Purpose:
# Identify duplicate records within the dataset.
#
# Duplicate observations may introduce bias into
# statistical analyses and machine learning models,
# potentially affecting the reliability of the results.
#
# ============================================================

# Count the number of duplicated rows
duplicates = df.duplicated().sum()

# Display the total number of duplicate records found
print("Number of duplicated rows:", duplicates)
```

```python
# ============================================================
# DUPLICATE RECORD REMOVAL
# ============================================================
#
# Purpose:
# Remove duplicate records from the dataset to ensure
# that each observation is unique.
#
# If no duplicate records are found, this operation
# will not modify the dataset.
#
# Removing duplicates helps prevent bias in statistical
# analyses and machine learning models by avoiding the
# overrepresentation of identical observations.
#
# ============================================================

# Remove duplicated rows from the dataset
df = df.drop_duplicates()

# Display the dataset dimensions after duplicate removal
print("Dataset shape after duplicate removal:", df.shape)
```

```python
# ============================================================
# SAVE DATASET TO SQLITE DATABASE
# ============================================================
#
# Purpose:
# Store the cleaned and prepared dataset in a SQLite
# database for persistent storage and future retrieval.
#
# SQLite was selected because it is lightweight,
# serverless, and well-suited for analytical projects
# developed within Jupyter Notebook environments.
#
# Database:
# - faostat_food_energy.db
#
# Table:
# - food_energy
#
# ============================================================

# Create a connection to the SQLite database
conn = sqlite3.connect("faostat_food_energy.db")

# Save the DataFrame as a database table
df.to_sql(
    "food_energy",
    conn,
    if_exists="replace",
    index=False
)

# Confirm successful storage
print("Dataset saved successfully to SQLite database.")
```

# Coding Regerence
https://www.sqlite.org/docs.html

```python
# ============================================================
# LOAD DATA FROM SQLITE DATABASE
# ============================================================
#
# Purpose:
# Retrieve the stored dataset from the SQLite database
# and load it into a Pandas DataFrame.
#
# This step verifies that the data has been successfully
# saved and demonstrates how datasets can be retrieved
# programmatically using SQL queries.
#
# Database:
# - faostat_food_energy.db
#
# Table:
# - food_energy
#
# ============================================================

# SQL query to retrieve all records from the database table
query = """
SELECT *
FROM food_energy;
"""

# Execute the SQL query and load the results into a DataFrame
df_from_db = pd.read_sql_query(query, conn)

# Display dataset dimensions to verify successful loading
print("Dataset shape:", df_from_db.shape)

# Display the first records for inspection
df_from_db.head()
```

```python
# Filtrar apenas Irlanda usando SQL
query = """
SELECT *
FROM food_energy
WHERE area = 'Ireland';
"""

ireland_db = pd.read_sql_query(query, conn)

print(ireland_db.shape)
ireland_db.head()
```

```python
#Agregar consumo energético por país e ano usando SQL

query = """
SELECT area, year, SUM(value) AS total_energy
FROM food_energy
GROUP BY area, year
ORDER BY area, year;
"""

country_year_energy_sql = pd.read_sql_query(query, conn)

country_year_energy_sql.head()
```

```python
# ============================================================
# RECORDS PER YEAR (SQL)
# ============================================================
# Purpose:
# Retrieve the number of records available for each year
# directly from the SQLite database.
# ============================================================

query = """
SELECT
    year,
    COUNT(*) AS count
FROM food_energy
GROUP BY year
ORDER BY year;
"""

records_year = pd.read_sql_query(query, conn)

print(records_year.head())

# ============================================================
# Visualization
# ============================================================

plt.figure(figsize=(8, 4))

sns.barplot(
    data=records_year,
    x="year",
    y="count"
)

plt.title("Records per Year")
plt.xlabel("Year")
plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.show()

```

```python
# ============================================================
# RECORDS PER COUNTRY (SQL)
# ============================================================
# Purpose:
# Count the number of observations available for each country
# stored in the database.
# ============================================================

query = """
SELECT
    area,
    COUNT(*) AS count
FROM food_energy
GROUP BY area
ORDER BY count DESC;
"""

records_country = pd.read_sql_query(query, conn)

print(records_country)

# ============================================================
# Visualization
# ============================================================

plt.figure(figsize=(8, 4))

sns.barplot(
    data=records_country,
    x="area",
    y="count"
)

plt.title("Records per Country")
plt.xlabel("Country")
plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.show()
```

```python
# ============================================================
# RECORDS PER FOOD CATEGORY (SQL)
# ============================================================
# Purpose:
# Count observations for each food system activity.
# ============================================================

query = """
SELECT
    item,
    COUNT(*) AS count
FROM food_energy
GROUP BY item
ORDER BY count DESC;
"""

records_item = pd.read_sql_query(query, conn)

print(records_item)

# ============================================================
# Visualization
# ============================================================

plt.figure(figsize=(8, 4))

sns.barplot(
    data=records_item,
    x="item",
    y="count"
)

plt.title("Records per Food System Activity")
plt.xlabel("Activity")
plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.show()
```

```python
# ============================================================
# RECORDS PER ENERGY SOURCE (SQL)
# ============================================================
# Purpose:
# Count observations available for each energy source.
# ============================================================

query = """
SELECT
    element,
    COUNT(*) AS count
FROM food_energy
GROUP BY element
ORDER BY count DESC;
"""

records_element = pd.read_sql_query(query, conn)

print(records_element)

# ============================================================
# Visualization
# ============================================================

plt.figure(figsize=(8, 4))

sns.barplot(
    data=records_element,
    x="element",
    y="count"
)

plt.title("Records per Energy Source")
plt.xlabel("Energy Source")
plt.ylabel("Number of Records")

plt.xticks(rotation=45)

plt.show()
```

```python
# ============================================================
# AVERAGE ENERGY CONSUMPTION BY COUNTRY (SQL)
# ============================================================
# Purpose:
# Calculate average energy consumption for each country.
# ============================================================

query = """
SELECT
    area,
    AVG(value) AS mean_energy
FROM food_energy
GROUP BY area
ORDER BY mean_energy;
"""

country_mean = pd.read_sql_query(query, conn)

print(country_mean)

# ============================================================
# Visualization
# ============================================================

plt.figure(figsize=(8, 5))

plt.scatter(
    country_mean["area"],
    country_mean["mean_energy"]
)

plt.title("Average Energy Consumption by Country")
plt.xlabel("Country")
plt.ylabel("Average Energy Consumption (TJ)")

plt.xticks(rotation=45)

plt.show()
```

```python
# ============================================================
# COUNTRY-YEAR ENERGY AGGREGATION (SQL)
# ============================================================
# Purpose:
# Aggregate total energy consumption by country and year.
# ============================================================

query = """
SELECT
    area,
    year,
    SUM(value) AS total_energy
FROM food_energy
GROUP BY area, year
ORDER BY area, year;
"""

country_year_energy = pd.read_sql_query(query, conn)

print(country_year_energy.head())
```

```python
# ============================================================
# AVAILABILITY MATRIX (SQL)
# ============================================================
# Purpose:
# Retrieve observations from the database and create a
# pivot table for visualization.
# ============================================================

query = """
SELECT
    item,
    element,
    value
FROM food_energy;
"""

availability_df = pd.read_sql_query(query, conn)

availability = pd.pivot_table(
    availability_df,
    index="item",
    columns="element",
    values="value",
    aggfunc="count"
)

print(availability)

# ============================================================
# Visualization
# ============================================================

plt.figure(figsize=(10, 6))

sns.heatmap(
    availability,
    annot=True,
    fmt=".0f",
    cmap="Blues"
)

plt.title("Availability of Observations by Activity and Energy Source")

plt.xlabel("Energy Source")
plt.ylabel("Food System Activity")

plt.show()
```

1. Intervalo de Confiança (Confidence Interval)
Objetivo

O intervalo de confiança foi utilizado para estimar a média populacional do consumo energético associado ao sistema alimentar da Irlanda. Enquanto a média amostral fornece apenas um valor pontual, o intervalo de confiança permite incorporar a incerteza estatística associada aos dados observados.

Fundamentação Teórica

Em estudos estatísticos, raramente é possível observar toda a população. Assim, utiliza-se uma amostra para estimar parâmetros populacionais. O intervalo de confiança representa uma faixa de valores plausíveis para o parâmetro de interesse (Montgomery and Runger, 2018).

Fórmula:

CI = x̄ ± tα/2 × (s/√n)

Onde:

x̄ = média amostral;
s = desvio padrão;
n = número de observações;
t = valor crítico da distribuição t de Student.

```python
import scipy.stats as stats

ireland_values = df[df["area"] == "Ireland"]["value"] 

mean_ireland = ireland_values.mean() 

sem = stats.sem(ireland_values) 

ci_ireland = stats.t.interval( 0.95, df=len(ireland_values)-1, loc=mean_ireland, scale=sem ) 

print(ci_ireland)
```

O intervalo de confiança de 95% estimou que o consumo energético médio do sistema alimentar irlandês situa-se entre 3405,56 TJ e 5117,45 TJ. Este resultado fornece uma medida de precisão para a média observada e reduz a incerteza associada às comparações internacionais realizadas neste estudo. A principal vantagem deste método é permitir inferências sobre a população sem necessidade de observar todos os dados disponíveis (Montgomery and Runger, 2018). Contudo, o intervalo não explica as causas das diferenças observadas nem identifica relações entre variáveis. Estudos relacionados utilizando dados agrícolas da FAO e sistemas de apoio à decisão agrícola também empregam intervalos de confiança para apoiar comparações entre países e validação de indicadores energéticos e produtivos (James et al., 2021; Osinga et al., 2022). Para este trabalho, o resultado estabelece uma referência estatística sólida para comparar a Irlanda com outros países da União Europeia e apoiar futuras etapas de previsão e Machine Learning.


**Relação com o Trabalho**

Como a Irlanda é o país de referência deste estudo, estimar a média populacional do seu consumo energético permite avaliar a precisão das conclusões obtidas e comparar esse comportamento com os restantes países europeus.

Trabalhos Relacionados

Estudos agrícolas e económicos frequentemente utilizam intervalos de confiança para avaliar a estabilidade de indicadores nacionais antes de comparações internacionais (James et al., 2021).

Referências

James, G., Witten, D., Hastie, T. and Tibshirani, R. (2021) An Introduction to Statistical Learning. 2nd edn. New York: Springer.

Montgomery, D.C. and Runger, G.C. (2018) Applied Statistics and Probability for Engineers. 7th edn. Hoboken: Wiley.

Osinga, S.A., Paudel, D., Mouzakitis, S.A. and Athanasiadis, I.N. (2022) ‘Big data applications in smart farming and agricultural decision support systems’, Agricultural Systems, 196.

James et al. (2021)

Montgomery and Runger (2018)

```python
# Confidence Interval - Mean with 95% Confidence Interval
import matplotlib.pyplot as plt

mean_ireland = ireland_values.mean()

plt.figure(figsize=(6,4))

plt.errorbar(
    x=["Ireland"],
    y=[mean_ireland],
    yerr=[[mean_ireland-ci_ireland[0]],
          [ci_ireland[1]-mean_ireland]],
    fmt='o',
    capsize=8
)

plt.ylabel("Energy Use (TJ)")
plt.title("Ireland Mean Energy Consumption with 95% Confidence Interval")

plt.show()
```

**2 . Teste de Shapiro-Wilk**

Objetivo

Verificar se os dados seguem uma distribuição normal.

Fundamentação Teórica

Muitos testes estatísticos paramétricos exigem normalidade dos dados. O teste de Shapiro-Wilk é amplamente utilizado para avaliar esse pressuposto (Field, 2018).

Hipóteses:

H₀: Os dados seguem distribuição normal.

H₁: Os dados não seguem distribuição normal.



**Um detalhe importante**

Esse intervalo ficou relativamente estreito:

3405 → 5117

o que é um bom sinal, pois indica que a estimativa da média da Irlanda possui uma precisão razoável.

Agora o próximo teste mais importante é o Shapiro-Wilk, porque ele vai determinar se podemos confiar em testes paramétricos (t-test e ANOVA) ou se devemos priorizar Mann-Whitney e Kruskal-Wallis.

```python
shapiro_test = stats.shapiro( ireland_values ) 

print(shapiro_test)
```

O teste de Shapiro-Wilk apresentou p-value = 6.12 × 10⁻¹², valor significativamente inferior a 0,05. Dessa forma, a hipótese nula foi rejeitada, indicando que os dados de consumo energético da Irlanda não seguem uma distribuição normal. Este resultado confirma as evidências observadas anteriormente no histograma e no boxplot, que mostraram forte assimetria positiva e presença de outliers. A principal vantagem desta conclusão é justificar o uso de métodos não paramétricos, como Mann-Whitney e Kruskal-Wallis, mais adequados para este conjunto de dados. Como limitação, a não normalidade pode reduzir a eficácia de alguns modelos estatísticos tradicionais. Estudos em análise agrícola e sistemas inteligentes de apoio à decisão também reportam distribuições não normais em indicadores energéticos e produtivos, justificando o uso de abordagens robustas e técnicas de Machine Learning (Field, 2018; Osinga et al., 2022). Para este trabalho, o resultado fornece uma base metodológica sólida para a escolha dos testes estatísticos subsequentes e dos algoritmos preditivos.

**Relação com o Trabalho**

O histograma e o boxplot desenvolvidos durante a EDA sugeriram forte assimetria positiva. O teste de Shapiro-Wilk foi utilizado para confirmar formalmente essa observação antes da escolha dos testes inferenciais.

**3 Teste de Levene**


Objetivo

Avaliar se a variabilidade do consumo energético da Irlanda é semelhante à de Portugal ou Grécia.

Fundamentação Teórica

A igualdade das variâncias é um pressuposto fundamental de diversos testes paramétricos. O teste de Levene é utilizado para verificar essa condição (Field, 2018).

Hipóteses:

H₀: As variâncias são iguais.

H₁: As variâncias são diferentes.

```python
portugal_values = df[ df["area"]=="Portugal" ]["value"] 

levene_test = stats.levene( ireland_values, portugal_values ) 

print(levene_test)
```

Relação com o Trabalho

Os gráficos comparativos mostraram que a dispersão dos dados varia significativamente entre os países. O teste de Levene foi aplicado para verificar se o uso de testes paramétricos tradicionais seria apropriado.

Referências

Field (2018)

Montgomery and Runger (2018)

O teste de Levene foi aplicado para verificar se as variâncias do consumo energético da Irlanda e de Portugal podem ser consideradas estatisticamente equivalentes. O resultado obtido apresentou estatística = 3,31 e p-value = 0,070, valor superior ao nível de significância de 0,05. Assim, não existem evidências suficientes para rejeitar a hipótese nula, sugerindo que as variâncias dos dois países podem ser consideradas homogêneas.

Portugal foi selecionado para esta comparação por apresentar características socioeconómicas e agrícolas mais próximas da Irlanda quando comparado com países como Alemanha, França ou Itália. Além disso, durante a Análise Exploratória dos Dados, os gráficos de tendência temporal mostraram níveis médios de consumo energético relativamente mais próximos dos observados na Irlanda, tornando a comparação mais relevante do ponto de vista estatístico e económico.

A principal vantagem deste resultado é que ele satisfaz um dos pressupostos necessários para a aplicação de testes paramétricos, aumentando a confiança nas análises comparativas subsequentes. Como limitação, o teste avalia apenas a igualdade das variâncias e não fornece informações sobre diferenças entre médias ou distribuições. Estudos comparativos em agricultura, energia e comércio internacional frequentemente utilizam o teste de Levene antes da aplicação de testes de comparação entre países, garantindo a validade dos pressupostos estatísticos (Field, 2018; Montgomery and Runger, 2018).

No contexto deste trabalho, o resultado indica que Irlanda e Portugal apresentam níveis semelhantes de variabilidade no consumo energético do sistema alimentar, reforçando a adequação de Portugal como país de referência para as análises comparativas desenvolvidas nas etapas seguintes.

**4. Welch's t-test**

Objetivo

Comparar o consumo energético médio da Irlanda com o de um país europeu semelhante.

Fundamentação Teórica

O Welch t-test é uma adaptação do Student t-test que não exige igualdade das variâncias.

Hipóteses:

H₀: As médias são iguais.

H₁: As médias são diferentes.

Fórmula:

        t = (x̄₁ − x̄₂) / √[(s₁²/n₁)+(s₂²/n₂)]

```python
welch_test = stats.ttest_ind( ireland_values, portugal_values, equal_var=False ) 

print(welch_test)
```

O teste de Welch foi utilizado para comparar o consumo energético médio da Irlanda e de Portugal, verificando se as diferenças observadas entre os dois países são estatisticamente significativas. O resultado apresentou estatística t = -0,917 e p-value = 0,359, valor superior ao nível de significância de 0,05. Assim, não existem evidências suficientes para rejeitar a hipótese nula, indicando que as médias de consumo energético dos dois países podem ser consideradas estatisticamente semelhantes.

A escolha de Portugal foi motivada pela sua maior proximidade económica e estrutural em relação à Irlanda quando comparada com economias maiores como Alemanha ou França. Além disso, a análise exploratória demonstrou que ambos os países apresentam níveis médios de consumo energético relativamente próximos e tendências temporais comparáveis, tornando a comparação mais adequada para este estudo.

A principal vantagem deste resultado é indicar que a Irlanda e Portugal possuem comportamentos energéticos semelhantes dentro do sistema alimentar analisado. Essa semelhança pode ser útil para futuras comparações de políticas agrícolas, eficiência energética e estratégias de sustentabilidade. Como limitação, o teste compara apenas as médias globais e não considera possíveis diferenças específicas entre atividades alimentares ou fontes energéticas.

Este resultado também complementa o intervalo de confiança calculado anteriormente e reforça as evidências observadas nos gráficos comparativos. **5.  Mann-Whitney U**

Objetivo

Comparar Irlanda e Portugal sem assumir normalidade.

Fundamentação Teórica

O Mann-Whitney U é um teste não paramétrico utilizado quando a distribuição dos dados não é normal.

Hipóteses:

H₀: As distribuições são iguais.

H₁: As distribuições são diferentes.

```python
mann_test = stats.mannwhitneyu(
    ireland_values,
    portugal_values,
    alternative="two-sided"
)

print(mann_test)
```

O teste de Mann-Whitney U foi aplicado como alternativa não paramétrica ao Welch's t-test, uma vez que o teste de Shapiro-Wilk demonstrou que os dados da Irlanda não seguem uma distribuição normal. Este teste compara as distribuições de duas populações sem assumir normalidade, sendo particularmente adequado para conjuntos de dados com assimetria e presença de outliers.

O resultado obtido apresentou estatística U = 9442 e p-value = 0,849, valor significativamente superior ao nível de significância de 0,05. Consequentemente, não existem evidências estatísticas suficientes para rejeitar a hipótese nula, indicando que as distribuições do consumo energético da Irlanda e de Portugal são estatisticamente semelhantes.

Este resultado é consistente com o Welch's t-test realizado anteriormente, que também não identificou diferenças significativas entre os dois países. A concordância entre um teste paramétrico e um teste não paramétrico fortalece a robustez das conclusões obtidas, aumentando a confiança nos resultados da análise.

A principal vantagem deste resultado é demonstrar que a semelhança observada entre Irlanda e Portugal não depende dos pressupostos de normalidade dos dados. Isso é particularmente relevante neste estudo, uma vez que o histograma e o boxplot revelaram forte assimetria positiva e presença de valores extremos. Como limitação, o teste não permite identificar quais fatores específicos explicam essa semelhança, nem considera separadamente as diferentes atividades do sistema alimentar ou fontes energéticas.

Diversos estudos em análise agrícola, energética e económica utilizam o teste de Mann-Whitney para comparar países ou regiões quando os dados apresentam distribuições não normais, situação comum em indicadores de produção, consumo energético e comércio internacional (Field, 2018; Osinga et al., 2022).

**5 . One-Way ANOVA**
    
Objetivo

Comparar simultaneamente os sete países analisados.

Fundamentação Teórica

A ANOVA permite verificar se existem diferenças estatisticamente significativas entre múltiplos grupos.

Hipóteses:

H₀: Todas as médias são iguais.

H₁: Pelo menos uma média é diferente.

Fórmula:

F = MSbetween / MSwithin

```python
groups = [
    group["value"].values
    for _, group in df.groupby("area")
]

anova_test = stats.f_oneway(*groups)

print(anova_test)
```

A análise de variância (ANOVA) foi utilizada para comparar simultaneamente os níveis médios de consumo energético dos sete países analisados: Irlanda, Portugal, Espanha, França, Alemanha, Itália e Grécia. Enquanto os testes t são adequados para comparações entre dois grupos, a ANOVA permite avaliar diferenças entre múltiplas populações de forma simultânea, reduzindo o risco de erros associados à realização de múltiplas comparações independentes (Montgomery and Runger, 2018).

O resultado obtido apresentou F = 46,13 e p-value = 5,31 × 10⁻⁵⁰, valor significativamente inferior ao nível de significância de 0,05. Dessa forma, a hipótese nula foi rejeitada, indicando que existem diferenças estatisticamente significativas entre os níveis médios de consumo energético observados nos países analisados.

Este resultado confirma as evidências observadas durante a Análise Exploratória dos Dados. Os gráficos de tendência temporal demonstraram que Alemanha, França e Itália apresentam níveis médios de consumo energético substancialmente superiores aos observados na Irlanda, Portugal e Grécia. A ANOVA fornece uma validação estatística formal dessas diferenças.

A principal vantagem deste teste é permitir uma comparação global entre todos os países analisados. No entanto, a ANOVA não identifica especificamente quais países diferem entre si, apenas indica que pelo menos um grupo apresenta comportamento distinto. Por esse motivo, testes complementares ou análises específicas entre países podem ser necessários para aprofundar a interpretação dos resultados.

Estudos relacionados em análise agrícola, consumo energético e sistemas alimentares frequentemente utilizam ANOVA para identificar diferenças entre regiões ou países antes da implementação de modelos preditivos e sistemas de apoio à decisão (James et al., 2021; Osinga et al., 2022).

No contexto deste trabalho, o resultado sugere que o consumo energético do sistema alimentar não é homogéneo entre os países europeus analisados. Esta conclusão reforça a importância de investigar fatores específicos associados a cada país e fornece suporte para a utilização de variáveis geográficas e nacionais nos modelos de Machine Learning desenvolvidos posteriormente.

Hipóteses

H₀: Todos os países possuem a mesma média de consumo energético.

H₁: Pelo menos um país possui média de consumo energético diferente.

Decisão Estatística

Como p-value = 5,31 × 10⁻⁵⁰ < 0,05, rejeita-se H₀.

Conclusão

Existem diferenças estatisticamente significativas entre os níveis médios de consumo energético dos países analisados. Este resultado confirma as diferenças observadas nos gráficos exploratórios e demonstra que o país é um fator relevante para explicar o consumo energético dentro do sistema alimentar europeu.

```python
ireland_yearly = (
    df[df["area"] == "Ireland"]
    .groupby("year")["value"]
    .sum()
    .reset_index()
)

spearman_test = stats.spearmanr(
    ireland_yearly["year"],
    ireland_yearly["value"]
)

print(spearman_test)
```

O teste de Correlação de Spearman retornou:

ρ (rho) = 0.945
p-value = 1.12 × 10⁻⁵

Este é provavelmente um dos resultados mais fortes de toda a análise estatística


A correlação de Spearman foi utilizada para avaliar a relação entre o tempo (2013–2023) e o consumo energético do sistema alimentar da Irlanda. Este teste foi escolhido porque os resultados do teste de Shapiro-Wilk demonstraram que os dados não seguem uma distribuição normal, tornando a correlação de Spearman mais adequada do que a correlação de Pearson (Field, 2018).

O resultado obtido apresentou um coeficiente de correlação de ρ = 0,945 e um p-value = 1,12 × 10⁻⁵, indicando uma correlação positiva extremamente forte e estatisticamente significativa entre o ano e o consumo energético da Irlanda. Como o valor de rho está muito próximo de 1, conclui-se que o consumo energético tende a aumentar consistentemente ao longo do período analisado.

Este resultado complementa os gráficos temporais desenvolvidos anteriormente, especialmente aqueles relacionados às atividades Food Retail e Food Processing, onde já era possível observar uma tendência de crescimento ao longo dos anos. A correlação de Spearman fornece uma validação estatística formal dessas observações visuais.

A principal vantagem deste resultado é demonstrar que existe uma tendência temporal clara no sistema alimentar irlandês. Essa informação é particularmente relevante para a próxima fase do projeto, que envolve previsão (forecasting) e Machine Learning, pois confirma a existência de um padrão temporal que pode ser explorado pelos modelos preditivos.

Como limitação, a correlação não implica causalidade. Embora exista uma associação forte entre o ano e o consumo energético, o teste não permite identificar quais fatores económicos, tecnológicos ou agrícolas são responsáveis por esse crescimento.

Essa justificativa é muito forte porque conecta diretamente a escolha das features aos resultados que você já produziu na EDA e na análise estatística.

```python
# ============================================================
# LOAD DATA FROM SQLITE
# ============================================================
#
# Purpose:
# Retrieve all records from the food_energy table.
#
# ============================================================

query = """
SELECT
    area,
    year,
    item,
    element,
    value
FROM food_energy
"""

df = pd.read_sql_query(query, conn)

print("Dataset shape:", df.shape)
df.head()
```

```python
# ============================================================
# 2. SELEÇÃO DE FEATURES E TARGET
# ============================================================

features = ["area", "year", "item", "element"]
target = "value"

X = df[features]
y = df[target]

print("Shape de X:", X.shape)
print("Shape de y:", y.shape)
```

**6 Separação treino e teste**

Os dados foram divididos em treino e teste para avaliar a capacidade dos modelos de generalizar para dados não vistos. Foi utilizada uma divisão de 80% para treino e 20% para teste. Essa etapa é essencial para evitar avaliar o modelo apenas nos dados usados durante o treinamento.

➡️ Como funciona a separação treino e teste

A separação treino/teste divide o dataset em duas partes:

✅ 80% → treino

✅ 20% → teste

O modelo aprende padrões usando apenas os dados de treino. Depois, ele é avaliado nos dados de teste, que ele ainda não viu. Isso permite verificar se o modelo realmente aprendeu padrões gerais ou apenas “decorou” os dados.

O que cada variavel significa:

* X = features: area, year, item, element
* y = target: value
* test_size=0.20 = 20% para teste
* random_state=42 = garante que o resultado seja reproduzível

Essas features foram utilizadas com o intuito de prever o consumo energético do sistema alimentar, o modelo precisa ser testado em dados não usados no treino. Isso ajuda a avaliar se ele consegue prever novos casos, por exemplo, novos anos, países ou combinações de atividade alimentar e fonte energética.

```python
# ============================================================
# 3. DIVISÃO ENTRE TREINO E TESTE
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Treino:", X_train.shape)
print("Teste:", X_test.shape)
```

A separação entre dados de treino e teste foi aplicada para avaliar a capacidade de generalização dos modelos preditivos. O conjunto de treino foi utilizado para ajustar os parâmetros dos algoritmos, enquanto o conjunto de teste foi reservado exclusivamente para avaliar o desempenho final. Essa abordagem reduz o risco de overfitting, situação em que o modelo apresenta bom desempenho nos dados de treino, mas baixa capacidade de prever dados não observados. Neste estudo, foi utilizada uma divisão de 80% para treino e 20% para teste, permitindo que os modelos fossem avaliados de forma mais realista no problema de previsão do consumo energético do sistema alimentar. Essa prática é amplamente recomendada em Machine Learning supervisionado (Hastie, Tibshirani and Friedman, 2009; James et al., 2021; Géron, 2022).

**7 Pré-processamento**


Como o dataset contém variáveis categóricas, foi aplicado One-Hot Encoding às colunas area, item e element. A variável year foi mantida como numérica. Para a Regressão Linear, a padronização pode melhorar a estabilidade do modelo. Para modelos baseados em árvores, como Random Forest e Gradient Boosting, a escala dos dados não é tão crítica, mas o mesmo pipeline foi mantido para consistência metodológica.



```python
# ============================================================
# 4. PRÉ-PROCESSAMENTO
# ============================================================

# Separação entre Variáveis Categóricas e Numéricas
categorical_features = ["area", "item", "element"]
numeric_features = ["year"]



preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),#Codificação das Variáveis Categóricas
        ("num", StandardScaler(), numeric_features) # Normalização da Variável Temporal
    ]
)
```


**8 Modelo 1 — Linear Regression**

A Regressão Linear foi utilizada como modelo base. Ela permite avaliar se uma abordagem simples é suficiente para explicar o consumo energético. No entanto, devido à assimetria, outliers e relações possivelmente não lineares identificadas na EDA, espera-se que modelos mais complexos tenham desempenho superior.

```python
# ============================================================
# 5. MODELO 1: LINEAR REGRESSION
# ============================================================

linear_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", LinearRegression())
    ]
)

linear_model.fit(X_train, y_train)

y_pred_linear = linear_model.predict(X_test)

linear_results = {
    "Model": "Linear Regression",
    "MAE": mean_absolute_error(y_test, y_pred_linear),
    "RMSE": np.sqrt(mean_squared_error(y_test, y_pred_linear)),
    "R2": r2_score(y_test, y_pred_linear)
}

linear_results
```

**9 Modelo 2 — Random Forest com GridSearchCV**


O Random Forest foi escolhido porque combina múltiplas árvores de decisão, reduzindo variância e melhorando a capacidade preditiva. Esse modelo é apropriado para o dataset porque os testes estatísticos indicaram não normalidade e outliers. Além disso, o Random Forest consegue capturar relações não lineares entre país, ano, atividade alimentar e fonte energética.

```python
# ============================================================
# 6. MODELO 2: RANDOM FOREST COM GRIDSEARCHCV
# ============================================================

rf_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", RandomForestRegressor(random_state=42))
    ]
)

rf_param_grid = {
    "model__n_estimators": [100, 200],
    "model__max_depth": [None, 5, 10],
    "model__min_samples_split": [2, 5],
    "model__min_samples_leaf": [1, 2]
}

rf_grid = GridSearchCV(
    rf_pipeline,
    rf_param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1
)

rf_grid.fit(X_train, y_train)

print("Best parameters:", rf_grid.best_params_)
print("Best CV score:", rf_grid.best_score_)

y_pred_rf = rf_grid.predict(X_test)

rf_results = {
    "Model": "Random Forest",
    "MAE": mean_absolute_error(y_test, y_pred_rf),
    "RMSE": np.sqrt(mean_squared_error(y_test, y_pred_rf)),
    "R2": r2_score(y_test, y_pred_rf)
}

rf_results
```

O Random Forest teve um desempenho extremamente superior ao da Regressão Linear.

Interpretação Geral

O modelo Random Forest apresentou um desempenho extremamente elevado, explicando aproximadamente 99,5% da variabilidade observada no consumo energético.

Esse resultado indica que praticamente toda a variação do consumo energético foi capturada pelo modelo.

**10 Modelo 3 — Gradient Boosting com GridSearchCV**

O Gradient Boosting foi utilizado como terceiro modelo por sua capacidade de construir árvores sequenciais que corrigem erros dos modelos anteriores. Ele é adequado para problemas tabulares e pode capturar padrões complexos. A utilização do GridSearchCV permite testar diferentes combinações de hiperparâmetros e selecionar a configuração com melhor desempenho.

```python
# ============================================================
# 7. MODELO 3: GRADIENT BOOSTING COM GRIDSEARCHCV
# ============================================================

gb_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", GradientBoostingRegressor(random_state=42))
    ]
)

gb_param_grid = {
    "model__n_estimators": [100, 200],
    "model__learning_rate": [0.05, 0.1],
    "model__max_depth": [2, 3, 4],
    "model__min_samples_split": [2, 5]
}

gb_grid = GridSearchCV(
    gb_pipeline,
    gb_param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1
)

gb_grid.fit(X_train, y_train)

print("Best parameters:", gb_grid.best_params_)
print("Best CV score:", gb_grid.best_score_)

y_pred_gb = gb_grid.predict(X_test)

gb_results = {
    "Model": "Gradient Boosting",
    "MAE": mean_absolute_error(y_test, y_pred_gb),
    "RMSE": np.sqrt(mean_squared_error(y_test, y_pred_gb)),
    "R2": r2_score(y_test, y_pred_gb)
}

gb_results
```

Os resultados dos dois modelos baseados em árvores indicam que o consumo energético do sistema alimentar pode ser previsto com elevada precisão utilizando informações relacionadas ao país, atividade alimentar, fonte energética e período temporal. A superioridade dos modelos não lineares sobre a Regressão Linear sugere que existem interações complexas entre essas variáveis, impossíveis de serem capturadas por modelos lineares simples. Assim, o problema proposto pelo estudo pode ser considerado adequadamente resolvido através de técnicas de Machine Learning supervisionado.

```python
#Feature Importance — Gradient Boosting
# Melhor modelo Gradient Boosting
best_gb = gb_grid.best_estimator_

# Modelo dentro do pipeline
gb_model = best_gb.named_steps["model"]

# Nomes das features após o pré-processamento
feature_names = best_gb.named_steps["preprocessor"].get_feature_names_out()

# Criar dataframe com importâncias
gb_importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": gb_model.feature_importances_
}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10, 6))

sns.barplot(
    data=gb_importance_df.head(15),
    x="Importance",
    y="Feature"
)

plt.title("Gradient Boosting - Top 15 Most Important Features")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

Os modelos foram avaliados usando três métricas: MAE, RMSE e R². O MAE mede o erro médio absoluto, o RMSE penaliza erros maiores e o R² indica a proporção da variância explicada pelo modelo. A comparação dessas métricas permite identificar qual modelo apresenta melhor equilíbrio entre precisão e capacidade explicativa.

**12 Cross-validation**

A validação cruzada foi utilizada para aumentar a confiabilidade dos resultados. Em vez de avaliar o modelo em uma única divisão treino-teste, o cross-validation divide os dados em múltiplas partes, treinando e testando o modelo várias vezes. Isso reduz o risco de conclusões dependentes de uma única separação dos dados.

```python
# ============================================================
# 10. CROSS-VALIDATION DO MELHOR MODELO
# ============================================================

best_model = rf_grid.best_estimator_

cv_scores = cross_val_score(
    best_model,
    X,
    y,
    cv=5,
    scoring="r2"
)

print("CV scores:", cv_scores)
print("Mean CV R2:", cv_scores.mean())
print("Standard deviation:", cv_scores.std())
```
